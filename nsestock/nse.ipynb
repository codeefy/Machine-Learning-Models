```python
import pandas as pd
df=pd.read_csv('nse_all_stock_data.csv')
df.head()
```

```python
l = {}

for i in df.columns:
    if i != 'Date':
        l[i]= df[i].mean() # mean of each column except Date column 
```

```python
# create a list of columns with null values >8000
null_cols = []
for i in df.columns:
    if df[i].isnull().sum() > 8000:
        null_cols.append(i)

print(null_cols)
```

```python
# create a new df with columns not in null_cols
df2 = df.drop(columns=null_cols)

# fill the remaining null values with the mean of the column not with date column
df2.fillna(l, inplace=True)
df2
```

```python
#  write code for info of the each column in df2
for i in df2.columns:
    print(df2[i].info()) # info of each column in df2

```

```python
# Perform basic statistical analysis on the stock prices and volumes
for i in df2.columns:
    print(df2[i].describe()) # describe of each column in df



```

```python
# now we have to perform eda on the data

# 1. check for null values
print(df2.isnull().sum())

```

```python
# 2. check for duplicate values
print(df2.duplicated().sum())


# 3. check for outliers
import seaborn as sns
import matplotlib.pyplot as plt
for i in df2.columns:
    sns.boxplot(df2[i])
    plt.show()
```
