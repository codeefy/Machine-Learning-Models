```python
# is there any way to check outlier in the data without visualization 
# perform except date column
for i in df2.columns:
    if i != 'Date':
        q1 = df2[i].quantile(0.25)
        q3 = df2[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr) 
        print(i, df2[(df2[i] < lower_bound) | (df2[i] > upper_bound)].shape[0]) # number of outliers in each column except Date column

```

```python
# store the index of the outliers in a list
l = df2[(df2[i] < lower_bound) | (df2[i] > upper_bound)].index
print(l)
print(len(l)) # number of outliers in the data  
```

```python
#remove the outliers from the data 
df2.drop(index=l, inplace=True)
print(df2.shape) # shape of the data after removing the outliers
```
