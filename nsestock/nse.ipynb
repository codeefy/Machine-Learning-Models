```python
import pandas as pd
df=pd.read_csv('nse_all_stock_data.csv')
df.tail()
```

```python
l = {}

for i in df.columns:
    if i != 'Date':
        l[i]= df[i].mean()
```

```python
# create a list of columns with null values >8000
null_cols = []
for i in df.columns:
    if df[i].isnull().sum() > 8000:
        null_cols.append(i)

print(null_cols)
```

```python
# create a new df with columns not in null_cols
df2 = df.drop(columns=null_cols)

# fill the remaining null values with the mean of the column not with date column
df2.fillna(l, inplace=True)
df2
```

```python
# remove rows with null values
df1=df.dropna()
df.describe()
```

```python
# create a list of those which has null values  greater than 8000
l=[]
for i in df.columns:
    if df[i].isnull().sum()>8000:
        l.append(i)
print(l)
print(len(l))
```

```python
# drop column with null values greater than 8000
df2=df.drop(l,axis=1)
df2.shape
```

```python
#replace null values with mean of the column except date column
for i in df2.columns:
    if i!='Date':
        df2[i]=df2[i].fillna(df2[i].mean())
```
