```python
import pandas as pd
df=pd.read_csv('nse_all_stock_data.csv')
df.head()
```

```python
l = {}

for i in df.columns:
    if i != 'Date':
        l[i]= df[i].mean() # mean of each column except Date column 
```

```python
# create a list of columns with null values >8000
null_cols = []
for i in df.columns:
    if df[i].isnull().sum() > 8000:
        null_cols.append(i)

print(null_cols)
```

```python
# create a new df with columns not in null_cols
df2 = df.drop(columns=null_cols)

# fill the remaining null values with the mean of the column not with date column
df2.fillna(l, inplace=True)
df2
```

```python
# count duplicate rows
df2.duplicated().sum()
df2.drop_duplicates(inplace=True)
df2.shape
```

```python
# check for outliers in the data without visualizing
for i in df2.columns:
    if i != 'Date':
        q1 = df2[i].quantile(0.25)
        q3 = df2[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 -(1.5 * iqr)
        upper_bound = q3 +(1.5 * iqr)
        print(i, df2[(df2[i] < lower_bound) | (df2[i] > upper_bound)].shape[0])
```

```python
# store the outlier in a list
outliers = []
for i in df2.columns:
    if i != 'Date':
        q1 = df2[i].quantile(0.25)
        q3 = df2[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 -(1.5 * iqr)
        upper_bound = q3 +(1.5 * iqr)
        outliers.append(df2[(df2[i] < lower_bound) | (df2[i] > upper_bound)].shape[0])
        print(i, df2[(df2[i] < lower_bound) | (df2[i] > upper_bound)].shape[0])
```
