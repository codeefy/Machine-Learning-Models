```python
import pandas as pd
df=pd.read_csv('nse_all_stock_data.csv')
df.head()
```

```python
l = {}

for i in df.columns:
    if i != 'Date':
        l[i]= df[i].mean() # mean of each column except Date column 
```

```python
# create a list of columns with null values >8000
null_cols = []
for i in df.columns:
    if df[i].isnull().sum() > 8000:
        null_cols.append(i)

print(null_cols)
```

```python
# create a new df with columns not in null_cols
df3 = df.drop(columns=null_cols)

# fill the remaining null values with the mean of the column not with date column
df3.fillna(l, inplace=True)
df3
```

```python
#null Values
df3.isnull().sum()
```

```python
# count duplicate rows
df3.duplicated().sum()
df3.drop_duplicates(inplace=True)
df3.shape
```

```python
# calculate index of outlier in each column
outliers = {}
for i in df3.columns:
    if i != 'Date':
        q1 = df3[i].quantile(0.25)
        q3 = df3[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound  = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)
        outliers[i] = df3[(df3[i] < lower_bound) | (df3[i] > upper_bound)].index
print(outliers)
```

```python
# i want to delete all rows which have outliers in any column
outliers_index = []
for i in outliers.values():
    for j in i:
        outliers_index.append(j)
outliers_index = list(set(outliers_index))
outliers_index
```

```python
df4=df3.drop(index=outliers_index)
df4.shape
```
