```python
import pandas as pd
df=pd.read_csv('nse_all_stock_data.csv')
df.head()
```

```python
l = {}

for i in df.columns:
    if i != 'Date':
        l[i]= df[i].mean() # mean of each column except Date column 
```

```python
# create a list of columns with null values >8000
null_cols = []
for i in df.columns:
    if df[i].isnull().sum() > 8000:
        null_cols.append(i)

print(null_cols)
```

```python
# create a new df with columns not in null_cols
df3 = df.drop(columns=null_cols)

# fill the remaining null values with the mean of the column not with date column
df3.fillna(l, inplace=True)
df3
```

```python
df3.info()
```

```python
df3.describe()
```

```python
#null Values
df3.isnull().sum()
```

```python
# count duplicate rows
df3.duplicated().sum()
df3.drop_duplicates(inplace=True)
df3.shape
```

```python
# calculate index of outlier in each column
outliers = {}
for i in df3.columns:
    if i != 'Date':
        q1 = df3[i].quantile(0.25)
        q3 = df3[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound  = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)
        outliers[i] = df3[(df3[i] < lower_bound) | (df3[i] > upper_bound)].index
print(outliers)
```

```python
outliers_to_drop = [x for x in outliers[i] if x in df3.index]
df3.drop(outliers_to_drop, inplace=True)
df3.shape
```

```python
df4=df3.copy()

```

```python
# calculate outliers in each column number of outliers
outliers = {}
for i in df4.columns:
    if i != 'Date':
        q1 = df4[i].quantile(0.25)
        q3 = df4[i].quantile(0.75)
        iqr = q3 - q1
        lower_bound  = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)
        outliers[i] = df4[(df4[i] < lower_bound) | (df4[i] > upper_bound)].shape[0]
#print(outliers)
# drop the columns with outliers
outliers_to_drop = [i for i in outliers if outliers[i] > 0]
df4.drop(columns=outliers_to_drop, inplace=True)
df4.shape 
```
